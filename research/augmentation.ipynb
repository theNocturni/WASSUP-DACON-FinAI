## 1. BERT augmentation

```python
sentence = "아버지가 오늘 아침에 부엌에 있었다가 갑자기 일어나시더니 슬그머니 안방으로 들어가셨다."
ratio = 0.15
print('raw  sentence :              ', sentence)
print('random_masking_replacement:  ', random_masking_replacement(sentence, ratio))
print('random_insertion_replacement:', random_masking_insertion(sentence, ratio))
```

## 4. 기존 EDA 방법 (KorEDA) - easy data augmentation
- 무작위 삭제 random_delete
- 무작위 스왑 random_swap
- 동의어 대체 synonym_replacement
- 무작위 삽입 random_insert

```python
sentence = "이순신은 조선 중기의 매우 뛰어난 전략가이자 용맹한 무신이다."
print('원본 :               ', sentence)
print('random_deletion:     ', " ".join(random_deletion(sentence.split(), 0.15)))
print('random_swap:         ', " ".join(random_swap(sentence.split(), 1)))
print('synonym_replacement: ', " ".join(synonym_replacement(sentence.split(), 1)))
print('random_insertion:    ', " ".join(random_insertion(sentence.split(), 1)))
```

```python
for i in range(10):
  print('원본 :               ', answer[i])
  print('random_deletion:     ', " ".join(random_deletion( answer[i].split(), 0.15)))
  print('random_swap:         ', " ".join(random_swap( answer[i].split(), 1)))
  print('synonym_replacement: ', " ".join(synonym_replacement( answer[i].split(), 1)))
  print('random_insertion:    ', " ".join(random_insertion( answer[i].split(), 1)))
  print()
```

## 5. 증강파일 저장

### 5.1

```python
## random_masking_insertion

raw_question = train_data['Question'].values.tolist()
augment_question = []
for i in range(len(raw_question)):
  aug = BERT_aug.random_masking_insertion(raw_question[i], ratio)
  augment_question.append(aug)

augment_data = train_data.copy()
augment_data['Question'] = augment_question

for idx, row in train_data.iterrows():
  if row['Question'] in augment_question:
    augment_data.drop(idx, axis=0, inplace=True)

augment_data.reset_index(drop=True, inplace=True)

augment_data.info()

# augment_data.to_csv('./bert_insert.csv')
```

```python
## eda_rs_question

raw_question = train_data['Question'].values.tolist()
augment_question = []
for i in range(len(raw_question)):
  aug = " ".join(random_swap(raw_question[i].split(), 1))
  augment_question.append(aug)

augment_data = train_data.copy()
augment_data['Question'] = augment_question

for idx, row in train_data.iterrows():
  if row['Question'] in augment_question:
    augment_data.drop(idx, axis=0, inplace=True)

augment_data.reset_index(drop=True, inplace=True)

augment_data.info()

# augment_data.to_csv('./random_swap.csv')
```

```python
## eda_ri_question

raw_question = train_data['Question'].values.tolist()
augment_question = []
for i in range(len(raw_question)):
  aug = " ".join(random_insertion(raw_question[i].split(), 1))
  augment_question.append(aug)

augment_data = train_data.copy()
augment_data['Question'] = augment_question

for idx, row in train_data.iterrows():
  if row['Question'] in augment_question:
    augment_data.drop(idx, axis=0, inplace=True)

augment_data.reset_index(drop=True, inplace=True)

augment_data.info()

# augment_data.to_csv('./random_insertion.csv')
```
